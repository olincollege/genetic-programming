# Visualization of the Iris Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Loading

```python
# Load the dataset
iris = pd.read_csv("data/Iris.csv")
iris.head()
```

## Plot dataset by dimensions 

```python
def plot_by_dimensions(part: str):
    """
    Plot the dimensions of the specified part of the iris flower.
    Params:
        part (str): The part of the iris flower to plot ('Sepal' or 'Petal').
    """
    sns.scatterplot(data=iris, x=f"{part}LengthCm", y=f"{part}WidthCm", hue="Species")
    plt.xlabel(f"{part} Length (cm)")
    plt.ylabel(f"{part} Width (cm)")
    plt.title(f"Species by {part} Dimensions")
    plt.plot()
```

```python
plot_by_dimensions("Sepal")
```

```python
plot_by_dimensions("Petal")
```
